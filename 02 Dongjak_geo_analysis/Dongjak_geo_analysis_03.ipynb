```python
# 필요한 모듈 설치
import pandas as pd
import numpy as np
import seaborn as sns
import folium
import json
from folium.plugins import MarkerCluster
```

```python
store = pd.read_csv("C:/Users/kjh96/Github/공모전/data/정제 데이터/상가유해시설_주소_fin.csv",encoding="CP949")
store.head(3)
```

# 유해시설

```python
# 지리정보 시각화
geo_df = store

# 지리정보 시각화 모듈 folium 사용
map = folium.Map(location = [geo_df["좌표정보(X)"].mean(), geo_df["좌표정보(Y)"].mean()], zoom_start=12)

for n in geo_df.index :
                # 팝업에 들어갈 텍스트 지정
                popup_name = geo_df.loc[n,'사업장명'] + '-' + geo_df.loc[n,'지번주소']

                
                # 표시 마커 설정
                folium.CircleMarker(
                location = [geo_df.loc[n,'좌표정보(X)'], geo_df.loc[n,'좌표정보(Y)']],
                radius=3,
                popup = popup_name,
                color = "red",
                fill = True,
                fill_color = "red"
                ).add_to(map)

map.save('index.html') # 버젼 문제로 한글깨짐 따라서 html 형식으로 추출 
map
```

# 국가시설

```python
faclity = pd.read_csv("C:/Users/kjh96/Github/공모전/data/정제 데이터/국가시설_주소_fin.csv",encoding="CP949")
faclity.head(3)
```

```python
faclity_bed = faclity[faclity["시설종류"].str.contains("배수시설|열원시설|공공하수도중 간선하수관|고등학교")]
faclity_bed.head(3)
```

# 유해시설 및 국가 시설 및 녹지

```python
fa = faclity_bed[["시설명","좌표정보(X)","좌표정보(Y)","지번주소"]]
fa["종류"] = "시설"
```

```python
st = store[["사업장명","좌표정보(X)","좌표정보(Y)","지번주소"]]
st.rename(columns={"사업장명":"시설명"},inplace = True)
st["종류"] = "상가"
```

```python
fa_st = pd.concat([fa,st])
fa_st = fa_st.reset_index(drop=True)
```

```python
green = pd.read_csv("C:/Users/kjh96/Github/공모전/data/정제 데이터/자연녹지지역_주소.csv",encoding="CP949")
green = green[["용도지역명2","좌표정보(X)","좌표정보(Y)","지번주소"]]
green["종류"] = "녹지"
green.rename(columns = {"용도지역명2":"시설명"}, inplace = True)
```

```python
df = pd.concat([fa_st,green])
df = df.reset_index(drop=True)
```

```python
df.head(3)
```

```python
# 지리정보 시각화
geo_df = df

# 지리정보 시각화 모듈 folium 사용
map = folium.Map(location = [geo_df["좌표정보(X)"].mean(), geo_df["좌표정보(Y)"].mean()], zoom_start=12 )

for n in geo_df.index :
                # 팝업에 들어갈 텍스트 지정
                popup_name = geo_df.loc[n,'시설명'] + '-' + geo_df.loc[n,'지번주소']
                if geo_df.loc[n,'종류'] == '시설' :
                    icon_color = 'blue'
                elif geo_df.loc[n,'종류'] == '녹지' :
                    icon_color = 'green'
                else :
                    icon_color = 'red'
                    
                
                # 표시 마커 설정
                folium.CircleMarker(
                location = [geo_df.loc[n,'좌표정보(X)'], geo_df.loc[n,'좌표정보(Y)']],
                radius=50,
                radiusUnit="Meters",
                popup = popup_name,
                color = icon_color,
                fill = True ,
                weight = 0.001,
                fill_color = icon_color
                ).add_to(map)

map.save('index.html') # 버젼 문제로 한글깨짐 따라서 html 형식으로 추출 
map
```
